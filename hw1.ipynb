```python
import numpy as np
from sklearn.model_selection import LeaveOneOut, GridSearchCV
from sklearn.base import BaseEstimator, ClassifierMixin, RegressorMixin

from sklearn.metrics import DistanceMetric, f1_score, make_scorer
from sklearn.utils.multiclass import unique_labels
```

```python
def uniform(u: np.ndarray) -> np.ndarray:
    return 0.5 * (np.abs(u) <= 1)

def triangular(u: np.ndarray) -> np.ndarray:
    return (1 - np.abs(u)) * (np.abs(u) <= 1)

def epanechnikov(u: np.ndarray) -> np.ndarray:
    return 0.75 * (1 - np.power(u, 2)) * (np.abs(u) <= 1)

def quartic(u: np.ndarray) -> np.ndarray:
    return 15 / 16 * np.power(1 - np.power(u, 2), 2) * (np.abs(u) <= 1)
```

```python
class FixedNWRegressor(BaseEstimator, RegressorMixin):
    def __init__(self, metric, kernel, h) -> None:
        super().__init__()
        self.metric = metric
        self.kernel = kernel
        self.h = h

    def fit(self, X, y):
        self.X_ = X
        self.y_ = y
        return self
    
    def predict(self, X, eps=1e-6):
        distances = self.metric.pairwise(X, self.X_)
        kernel_res = self.kernel(distances / self.h)
        return (kernel_res @ self.y_) / (kernel_res.sum(axis=1) + eps)
```

```python
class KNeighNWRegressor(BaseEstimator, RegressorMixin):
    def __init__(self, metric, kernel, k) -> None:
        super().__init__()
        self.metric = metric
        self.kernel = kernel
        self.k = k

    def fit(self, X, y):
        self.X_ = X
        self.y_ = y
        return self
    
    def predict(self, X, eps=1e-6):
        distances = self.metric.pairwise(X, self.X_)
        distances.sort(axis=1)
        k_distance = distances[:, [self.k + 1]]
        kernel_res = self.kernel(distances / k_distance)
        return (kernel_res @ self.y_ ) / (kernel_res.sum(axis=1) + eps)
```

```python
class MultiClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, estimator_class, kernel=None, metric=None, h=None, k=None) -> None:
        super().__init__()
        self.metric = metric
        self.kernel = kernel
        self.h = h
        self.k = k
        self.estimator_class = estimator_class

    def fit(self, X, y):        
        self.classes_ = unique_labels(y)
        one_hot_y = pd.get_dummies(y, dtype=float)
        if self.h:
            self.estimators = {label: self.estimator_class(self.metric, 
                                                           self.kernel,
                                                           self.h).fit(X, one_hot_y[label])
                                                           for label in self.classes_}
        elif self.k:
            self.estimators = {label: self.estimator_class(self.metric, 
                                                           self.kernel,
                                                           self.k).fit(X, one_hot_y[label])
                                                           for label in self.classes_}
        else:
            raise NotImplementedError('No window type set!')
        self.y_ = y
        return self
    
    def predict(self, X):
        pred_values = np.array([self.estimators[label].predict(X) for label in self.classes_])
        pred_labels = np.where(np.argmax(pred_values, axis=0), *self.classes_)
        return pred_labels
```

```python
cv_splitter = LeaveOneOut()
metrics = ['euclidean', 'manhattan', 'chebyshev']
r_max = [np.max(DistanceMetric.get_metric(metric).pairwise(X)) for metric in metrics]
f1 = make_scorer(f1_score, labels=unique_labels(y), pos_label='P')
d = int(np.sqrt(X.shape[0]))
r_max, d, f1, cv_splitter
```

```python
results = []

for metric, max_distance in zip(metrics, r_max):
    param_grid = {'h': np.arange(max_distance / d, max_distance, max_distance / d),
                'metric': [DistanceMetric.get_metric(metric)],
                'kernel': [uniform, triangular, epanechnikov, quartic]}
    
    regressor = GridSearchCV(estimator=MultiClassifier(FixedNWRegressor), 
                            param_grid=param_grid, cv=cv_splitter,
                            scoring=f1, error_score='raise', n_jobs=8)
    regressor.fit(X, y)
    results.append(regressor)
```

```python
param_grid = {'k': np.arange(2, d),
            'metric': [DistanceMetric.get_metric(metric) for metric in metrics],
            'kernel': [uniform, triangular, epanechnikov, quartic]}
    
search_k = GridSearchCV(estimator=MultiClassifier(KNeighNWRegressor), 
                        param_grid=param_grid, cv=cv_splitter,
                        scoring=f1, error_score='raise', n_jobs=8)
search_k.fit(X, y)
```

```python
[(search.best_score_, search.best_params_) for search in results + [search_k]]
```

```python
model = MultiClassifier(FixedNWRegressor)
best_param_grid = {
    'h': np.arange(max_distance / d, max_distance, max_distance / d),
    'metric': [DistanceMetric.get_metric('euclidean')],
    'kernel': [quartic],
}

grid = GridSearchCV(model, cv=5, param_grid=best_param_grid)
grid.fit(X, y), grid.best_params_, grid.best_score_
```

```python
final_cv_results = pd.DataFrame(grid.cv_results_)
final_cv_results['param_h'] = final_cv_results['param_h'].astype(float)
final_cv_results[['param_h', 'mean_test_score']].plot(x='param_h', y='mean_test_score')
```

```python
model = MultiClassifier(KNeighNWRegressor)
best_param_grid = {
    'metric': [DistanceMetric.get_metric('euclidean')],
    'kernel': [quartic],
    'k': np.arange(1, d)
}

grid = GridSearchCV(model, cv=5, param_grid=best_param_grid)
grid.fit(X, y), grid.best_params_, grid.best_score_
```

```python
final_cv_results = pd.DataFrame(grid.cv_results_)
final_cv_results['param_k'] = final_cv_results['param_k'].astype(float)
final_cv_results[['param_k', 'mean_test_score']].plot(x='param_k', y='mean_test_score')
```
